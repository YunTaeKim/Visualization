## 1. 여러가지 그래프의 종류
### 1.1 막대그래프
*다른 범주로 잘 나눠진 스치 데이터를 다룰 때 효과적인 그래프.*

 - 좀 더 효과를 주기 위해 막대에 색을 추가한다. 하지만, 너무 많은 막대가 있는 경우, 많은 색이 그래프를 어설프게 보이게 할 수 있다.  
 - 대시보드에 여러 막대 그래프를 나타낸다.
 - 축 양쪽에 막대를 놓는다. 양수와 음수 데이터 모드를 연결된 축에 놓는 것이 경향을 파악하는 데 효과적이다.
 - 적층 막대나 나란한 막대를 사용한다. 
 
### 1.2 원 그래프
*"부분들이 의미있는 전체를 구성하는가" 그리고 "원형을 사용해 표현하기에 충분한 공간이 되는가"를 먼저 고려하여 그린다면,  
한 공간이나 지도에 걸쳐 비율을 보여주는 데 유용한 그래프.*
 
 - 원 그래프의 조각을 최대 8개로 제한한다.
 - 지도 위에 원 그래프를 겹친다.

### 1.3 상자 그래프
*데이터의 분포를 최솟값, 제 1사분위, 중앙값, 제 3 사분위, 최댓값의 5개 요약된 숫자에 기반해 표준화해 보여주는 그래프.*

 - 데이터 내의 이상점을 발견해 낸다.
 - 데이터가 양단으로 어떻게 편향돼 있는지 보여준다.
 - 상자 안의 점들을 숨긴다. 이상점에 집중한다.
 - 분포 전반을 비교한다. 상자 그래프는 데이터 집합 간의 분포를 빠르게 비교하는 데 유용하다.
 
### 1.4 산포도
*두 변수를 시각적으로 표현하는 그래프.*

 - 추세선 혹은 최적의 선(선형 관계라면)을 추가하여 상관성을 보여줄 수 있다.
 - 유용한 기호를 사용한다. 상대적 모양과 색으로 시각적 효과를 증대시켜 데이터에 관한 이야기를 잘 드러낼 수 있다.
 
### 1.5 버블 차트
*데이터의 3차원을 표현하며, 세 번째 데이터 속성이 두 축을 따라 어떻게 집중되는지 보여주는 데 유용한 그래프.*

 - 유용한 색과 기호를 추가한다. 크기와 색을 다양하게 함으로써, 어떤 질문들에 대한 명환학 답변을 줄 수 있다.
 - 인터렉티브하게 만든다. 너무 많은 데이터 점들이 있다면 군집화할 수 있다. 타임 축 또는 범주 값으로 그룹을 생성하고  
   인터랙티브하게 단계별로 시각화한다.
   
### 1.6 KDE 도표
*확률 밀도 함수를 추정하고, 완만한 근사치를 만들기 위해 데이터 점들의 평균을 구하는 커널밀도추정을 이용한 그래프.*

## 3. 인터랙티브 시각화
 시각화를 위해서는 상호작용이 반드시 고려돼야 하며, 다음 두가지 기준을 충족해야 한다.  
  - **사람의 입력**: 정보를 시각적으로 표현하는 일부 제어 권한은 사람이 조절할 수 있어야 한다.
  - **반응시간**: 사람에 의해 변경된 부분은 적절한 시간 후에 반드시 시각화 정보에 반영돼야 한다.
  
### 3.1 이벤트 리스너
*마우스가 움직이거나 클릭됐을 때 사용되는 프로세스.*
 - 인터랙티브 시각화를 할 때, 사용자는 확실히 행동이 완료됐다고 확인해 주어야 한다. 따라서, 선택된 아이템에 대한 강조와 같은  
   **시각적 피드백**이 필요하며, 시각적 피드백은 보통 1초 내에 일어나는 즉각적인 반응이어야 한다.
 - 하나의 동작이 사용자가 예상되는 것보다 오랫동안 일어날 경우, 처리되고 있다는 표시가 상요자에게 보여야 한다.
 
### 3.2 레이아웃
데이터를 시각적이고 효과적으로 표현하기 위해서는 레이아웃 방법을 아는 것이 중요하다. 레이아웃의 결과가 높은 가독성을  
갖도록 만들기 위해 가능하다면 계층이나 대칭을 가지는 구조가 필요하다. 하나의 중요한 요소는 공간의 효율화이다.
 - 원형 레이아웃 (Circular layout)
   - Link in a circle
   
   
 - 방사형 레이아웃 (Radial layout)
   - Sunburst plot
   - Treemap
   
   
 - 풍선 레이아웃 (Balloon layout)
   - Pie
   - Bubble
   
## 4. 파이썬의 성능 향상
### 4.1 Numbapro
 *파이썬 코드를 CUDA가능 CPUs 또는 멀티코어 CPU에서 컴파일할 수 있는 컨틴머 애널리틱스의 파이썬 컴파일러를 사용할 경우  
  중간 해석코드보다 몇배 더 빠르게 작동한다.*
  
### 4.2 Scipy.weave
 *C코드를 중간에 삽입하고, NumPy의 배열을 C층으로 자연스럽게 이동시켜주는 모듈로 효율적인 매크로를 가지고 있다.*
 
### 4.3 멀티코어 접근 방법
 *파이썬의 멀티 프로세싱 패키지는 서브 프롯스를 생성하는 상대적으로 단순한 메커니즘을 제공한다. 어떤 작업을 하기 위해서  
  모든 프로세서들을 사용하는 것이 빠르기 때문에 멀티코어 프로세서를 모두 활용하는 것이 합당하다.*
  
### 4.4 프로세스 풀(pool)
 *풀은 멀티프로세싱의 패키지에 또 다른 클래스로, 풀 안에 생성된 프로세스 수를 정의할 수 있고 각 프로세스에게  
  전달될 파라미터를 가지고 있는 반복 객체를 보낼 수 있다.*
  
### 4.5 분산 계산 패키지(ex. Disco)
 *분산 계산을 위해 생성된 MapReduce 패러다임 기반 가벼운 오픈 소스 프레임워크를 사용한다.*

```python
N = 7
winnersplot = (142.6, 125.3, 26.0, 81.0, 145.6, 319.4, 178.1)

ind = np.arange(N)

width = 0.35

fig, ax = plt.subplots(figsize = (14, 10))
winners = ax.bar(ind, winnersplot, width, color = '#ffad00')

nomineesplot = (109.4, 94.8, 60.7, 44.6, 116.9, 262.5, 102.0)
nominees = ax.bar(ind + width, nomineesplot, width, color = '#9b3c38')

ax.set_xticks(ind + width)
ax.set_xticklabels(('Best Picture', 'Directors', 'Best Actor', 'Best Actress', 'Editing', 'Visual Effects', 'Cinematography'))
ax.legend((winners[0], nominees[0]), ('Academy Award Winners', 'Academy Award Nominees'))

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        hcap = "$" + str(height) + 'M'
        ax.text(rect.get_x() + rect.get_width() / 2., height, hcap, ha = 'center', va = 'bottom', rotation = 'vertical')
        
autolabel(winners)
autolabel(nominees)

plt.show()
```

```python
eboladata = pd.read_csv('data/ebola.csv')
filtered = eboladata[eboladata['value'] > 0]
filtered = filtered[filtered['Indicator'] == 'Cumulative number of confirmed Ebola deaths']
sorteddata = filtered.sort_values('Date')

sierradata = sorteddata.loc[sorteddata['Country'] == 'Sierra Leone']
guineadata = sorteddata.loc[sorteddata['Country'] == 'Guinea']
liberiadata = sorteddata.loc[sorteddata['Country'] == 'Liberia']

g_x = [dt.datetime.strptime(row, '%Y-%m-%d').date() for row in guineadata['Date']]
g_y = list(guineadata['value'])
s_x = [dt.datetime.strptime(row, '%Y-%m-%d').date() for row in sierradata['Date']]
s_y = list(sierradata['value'])
l_x = [dt.datetime.strptime(row, '%Y-%m-%d').date() for row in liberiadata['Date']]
l_y = list(liberiadata['value'])

plt.figure(figsize = (10, 10))
plt.plot(g_x, g_y, color = 'red', linewidth = 2, label = 'Guinea')
plt.plot(s_x, s_y, color = 'orange', linewidth = 2, label = 'Sierra Leone')
plt.plot(l_x, l_y, color = 'blue', linewidth = 2, label = 'Liberia')

plt.xlabel('Date', fontsize = 18)
plt.ylabel('Number of Ebola Deaths', fontsize = 18)
plt.title('Confirmed Ebola Deaths', fontsize = 20)
plt.legend(loc = 2)
plt.show()
```

```python
def num(s):
    try:
        return(int(s)) 
    except ValueError:
        return(0)  

def getcolors():
    colors = [(31, 119, 180), (255,0,0), (0,255,0), (148, 103, 189), (140, 86, 75), (218, 73, 174), (127, 127, 127), (140,140,26), (23, 190, 207), (65,200,100), (200, 65,100), (125,255,32), (32,32,198), (255,191,201), (172,191,201), (0,128,0), (244,130,150), (255, 127, 14), (128,128,0), (10,10,10), (44, 160, 44), (214, 39, 40), (206,206,216)]

    for i in range(len(colors)):
        r, g, b = colors[i]
        colors[i] = (r / 255. , g / 255. , b / 255.)
    return(colors) 
```

```python
data = pd.read_csv('data/qb_data.csv')
qbnames = list(data['Name'].drop_duplicates())

rank = 0
highrank = 244
colindex = 0
lastage = 20

colorsdata = getcolors()

plt.figure(figsize = (15, 13))
plt.ylim(10, 744)
plt.xlim(1940, 2021)

for qbn in qbnames:    
    temp = data.loc[(data['Name'] == qbn) & (data['Year'] != 'Career')]
    x = []
    y = []
    prevsum = 0
    for i in temp.index:
        lastyr = num(temp.loc[i, 'Year'])
        lastage = num(temp.loc[i, 'Age'])
        prevsum += num(temp.loc[i, 'TD'])
        x += [num(temp.loc[i, 'Year'])]
        y += [prevsum]
            
    if (rank > highrank):
        plt.plot(x, y, color = colorsdata[colindex], label = qbn, linewidth = 2.5)
        plt.legend(loc = 0, prop = {'size': 10})
        colindex = (colindex + 1) % 22
        plt.text(lastyr - 1, prevsum + 2, qbn + '(' + str(prevsum) + '):' + str(lastage), fontsize = 9)

    else:
        plt.plot(x, y, color = colorsdata[22], linewidth = 1.5)
        rank = rank + 1

plt.xlabel('Year', fontsize = 18)
plt.ylabel('Cumulative Touch Downs', fontsize = 18)
plt.title('Cumulative Touch downs by Quarter Backs', fontsize = 20)
plt.show()
```

```python
sports = pd.read_csv('data/nfl_teams.csv')
fig = plt.figure(figsize = (15, 10), facecolor = 'w')

def plotCircle(x, y, radius, color, alphaval):
    circle = plt.Circle((x, y), radius = radius, fc = color, alpha = alphaval) # 테두리 없이 내부만 색이 있는 원
    fig.gca().add_patch(circle)
    nofcircle = plt.Circle((x, y), radius = radius, ec = color, fill = False) # 테두리만 있는 원
    fig.gca().add_patch(nofcircle)

x = sports[' Years Completed']
y = sports[' Num Championships']
r = sports[' Championships/yr average']
for i in range(0, len(x)):
    plotCircle(x[i], y[i], r[i], 'b', 0.1)

plt.axis('scaled')
plt.show()
```
